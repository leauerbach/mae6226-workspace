# Lesson 5: Infinite Row of Vortices

## Task 1

```python
import numpy
import math
from matplotlib import pyplot

%matplotlib inline
%run ./definitions.ipynb
```

```python
n = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -0.5, 0.5
x, y, X, Y = create_meshgrid(n,x_start,x_end,y_start,y_end)
```

```python
n_vortices = 151
x_vortices = numpy.linspace(-n_vortices/5,n_vortices/5,n_vortices)
y_vortices = numpy.zeros(n_vortices)
```

```python
gamma = 10
u_total = 0
v_total = 0
psi_total = 0

for i in range(numpy.size(x_vortices)):
    u, v = get_velocity_vortex(gamma,x_vortices[i],y_vortices[i],X,Y)
    u_total += u
    v_total += v
    psi = get_stream_function_vortex(gamma,x_vortices[i],y_vortices[i],X,Y)
    psi_total += psi
```

```python
size = 10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x',fontsize=16)
pyplot.ylabel('y',fontsize=16)
pyplot.xlim(-2.0,2.0)
pyplot.ylim(-0.5,0.5)
pyplot.streamplot(X,Y,u_total,v_total,density=1,linewidth=1,arrowsize=1,arrowstyle='->')
pyplot.scatter(x_vortices,y_vortices,color='#CD2305',s=80,marker='o');
```
